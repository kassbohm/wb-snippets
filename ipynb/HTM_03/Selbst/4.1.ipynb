```python
# https://colab.research.google.com/github/kassbohm/wb-snippets/blob/master/ipynb/HTM_03/Selbst/4.1.ipynb

from sympy import *

m = var("m", positive = True)
a, b, c = var("a, b, c", positive=True)
a2, b2, c2 = a*a, b*b, c*c

# Start Check Parallel Axis:
pprint("\nJ with respect to O:")
x, y, z = var("x, y, z")
V = a*b*c
pprint("11-Component:")
# J11
# = int ρ (y² + z²) dV
# = ρ int (y² + z²) dV
# = m / V int (y² + z²) dV
tmp = y*y + z*z
tmp = integrate(tmp, (x, 0, a))
tmp = integrate(tmp, (y, 0, b))
tmp = integrate(tmp, (z, 0, c))
tmp *= m / V
tmp = tmp.simplify()
pprint(tmp)
pprint("12-Component:")
# J12
# = - int ρ (xy) dV
tmp = -x*y
tmp = integrate(tmp, (x, 0, a))
tmp = integrate(tmp, (y, 0, b))
tmp = integrate(tmp, (z, 0, c))
tmp *= m / V
tmp = tmp.simplify()
pprint(tmp)

# End 903126 ---

J11 = m * (b2 + c2)
J22 = m * (a2 + c2)
J33 = m * (a2 + b2)

pprint("\nK-Components of JC:")
J = Matrix([
    [J11, 0, 0],
    [0, J22, 0],
    [0, 0, J33]
    ])/12
pprint(J)

pprint("\nK-Components of JC for b=a and c=a:")
Jabc = J.subs([(b,a), (c,a)])
pprint(Jabc)

pprint("\nK-Components of JO:")
JParallelAxis = m/4 * Matrix([
    [b2+c2,  -a*b,   -a*c],
    [-a*b,  a2+c2,   -b*c],
    [-a*c,   -b*c,  a2+b2]
    ])
JO = J + JParallelAxis
JO = JO.applyfunc(simplify)
pprint(JO)

pprint("\nK-Components of JO for b=a and c=a:")
JOabc = JO.subs([(b,a), (c,a)])
pprint(JOabc)

pprint("\nUnitvector along OE:")
dOE = Matrix([a, b, c])
d = dOE / dOE.norm()
pprint(d)

pprint("\nComponent along OE:")
JOE = d.transpose()*J*d
JOE = JOE.applyfunc(simplify)
pprint(JOE)

pprint("\nComponent along OE for b=a and c=a:")
JOE = d.transpose()*Jabc*d
JOE = JOE.applyfunc(simplify)
pprint(JOE[0])

# Maier, Dimitri:
pprint("\nR = R₂ R₁:")
C = sqrt(2)/2
S = sqrt(2)/2
R1 = Matrix([
    [C, 0, -S],
    [0, 1,  0],
    [S, 0,  C]
    ])
R2 = Matrix([
    [1,  0,  0],
    [0,  C,  S],
    [0, -S,  C]
    ])
R = R2*R1
pprint(R)

pprint("\nVar. 1 leads to correct result:")
pprint("\nM-Components of JC:")
JCM = R*Jabc*R.transpose()
pprint(JCM)
pprint("\nM-Components of JO:")
tmp = (sqrt(2)*a/2)**2
JOM = JCM + m*Matrix([
    [tmp, 0, 0],
    [0, 0, 0],
    [0, 0, tmp],
    ])
pprint(JOM)

pprint("\nVar. 2 leads to *wrong* result:")
pprint("\nM-Components of JO:")
JOM = R*JO*R.transpose()
JOMabc = JOM.subs([(b,a), (c,a)])
JOMabc = JOMabc.applyfunc(simplify)
pprint(JOMabc)
```
