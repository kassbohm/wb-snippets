```python
from sympy import *

m = var("m", positive = True)
a, b, c = var("a, b, c", positive=True)
a2, b2, c2 = a*a, b*b, c*c

J11 = m * (b2 + c2)
J22 = m * (a2 + c2)
J33 = m * (a2 + b2)

pprint("\nJC: Components wrt (e1, e2, e3):")
J = Matrix([
    [J11, 0, 0],
    [0, J22, 0],
    [0, 0, J33]
    ])/12

pprint(J)

pprint("\nd = Unitvector along OE:")
dOE = Matrix([a, b, c])
d = dOE / dOE.norm()
pprint(d)


pprint("\nJ = Component along OE:")
J = d.transpose()*J*d
J = J.applyfunc(simplify)
pprint(J)

# JC: Components wrt (e1, e2, e3):
# ⎡  ⎛ 2    2⎞                          ⎤
# ⎢m⋅⎝b  + c ⎠                          ⎥
# ⎢───────────       0            0     ⎥
# ⎢     12                              ⎥
# ⎢                                     ⎥
# ⎢               ⎛ 2    2⎞             ⎥
# ⎢             m⋅⎝a  + c ⎠             ⎥
# ⎢     0       ───────────       0     ⎥
# ⎢                  12                 ⎥
# ⎢                                     ⎥
# ⎢                            ⎛ 2    2⎞⎥
# ⎢                          m⋅⎝a  + b ⎠⎥
# ⎢     0            0       ───────────⎥
# ⎣                               12    ⎦
#
# d = Unitvector along OE:
# ⎡        a        ⎤
# ⎢─────────────────⎥
# ⎢   ______________⎥
# ⎢  ╱  2    2    2 ⎥
# ⎢╲╱  a  + b  + c  ⎥
# ⎢                 ⎥
# ⎢        b        ⎥
# ⎢─────────────────⎥
# ⎢   ______________⎥
# ⎢  ╱  2    2    2 ⎥
# ⎢╲╱  a  + b  + c  ⎥
# ⎢                 ⎥
# ⎢        c        ⎥
# ⎢─────────────────⎥
# ⎢   ______________⎥
# ⎢  ╱  2    2    2 ⎥
# ⎣╲╱  a  + b  + c  ⎦
#
# J = Component along OE:
# ⎡  ⎛ 2  2    2  2    2  2⎞⎤
# ⎢m⋅⎝a ⋅b  + a ⋅c  + b ⋅c ⎠⎥
# ⎢─────────────────────────⎥
# ⎢       ⎛ 2    2    2⎞    ⎥
# ⎣     6⋅⎝a  + b  + c ⎠    ⎦
```
