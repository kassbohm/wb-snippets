```python
# https://colab.research.google.com/github/kassbohm/wb-snippets/blob/master/ipynb/TEM_11/Selbst/2.7.ipynb

from sympy import *

B, q, l = var("B, q, l")
C1, C2 = var("C1, C2")

xi = var("xi")

x1 = xi**1
x3 = xi**3
x4 = xi**4

pprint("\nM:")
M   = q*l**2 * ( x1/3 - 2*x3/3 + x4/3 )
pprint(M)

pprint("\nw:")
wpp = - M / B
wp = integrate(wpp*l, xi) + C1
w  = integrate(wp*l, xi) + C2
pprint(w)

eq1 = Eq( w.subs(xi,0), 0)
eq2 = Eq( w.subs(xi,1), 0)

sol = solve([eq1, eq2],[C1, C2])

C1s = sol[C1]
C2s = sol[C2]

pprint("\n\nC1:")
pprint(C1s)

pprint("\n\nC2:")
pprint(C2s)

w = w.subs({C1: C1s, C2: C2s})
w = w.simplify()

pprint("\n\n90 B w / (ql⁴):")
pprint(w * 90*B/(q*l**4))

# M:
#      ⎛ 4      3    ⎞
#  2   ⎜ξ    2⋅ξ    ξ⎟
# l ⋅q⋅⎜── - ──── + ─⎟
#      ⎝3     3     3⎠
#
# w:
#                4    6    4    5    4    3
#               l ⋅q⋅ξ    l ⋅q⋅ξ    l ⋅q⋅ξ
# C₁⋅l⋅ξ + C₂ - ─────── + ─────── - ───────
#                 90⋅B      30⋅B      18⋅B
#
#
# C1:
#  3
# l ⋅q
# ────
# 30⋅B
#
#
# C2:
# 0
#
#
# 90 B w / (ql⁴):
#   ⎛   5      4      2    ⎞
# ξ⋅⎝- ξ  + 3⋅ξ  - 5⋅ξ  + 3⎠
```
